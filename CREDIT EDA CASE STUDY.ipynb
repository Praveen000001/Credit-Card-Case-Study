### Age

```python
plt.figure(figsize = (15,7))
plt.xlabel('AGE')
plt.ylabel('FREQUENCY')
credit_imp.AGE.hist(ec='blue')
plt.show()
```

```python
# Create Bins of Age
credit_imp['AGE_GROUP'] = pd.cut(credit_imp.AGE,[0,30,60,70],labels = ['YOUNG','MIDDLEAGE', 'OLD'])
round(credit_imp.AGE_GROUP.value_counts(normalize=True) * 100, 2)
```

```python
# Plot histogram of AGE and Bar Plot of AGE_GROUP to view binning

plt.figure(figsize = (15,7))

plt.subplot(121, title='AGE Histogram Distribution')
plt.xlabel('AGE')
plt.ylabel('FREQUENCY')
credit_imp.AGE.hist(ec='blue', bins=15)

plt.subplot(122, title='AGE_GROUP Bar Plot')
plt.xlabel('AGE_GROUP')
plt.ylabel('PERCENTAGE')
(credit_imp.AGE_GROUP.value_counts(normalize=True)*100).plot.bar(rot=0)

plt.show()
```

```python
## AGE_GROUP column has been created from AGE column using Binning.
```

```python
# Binning for Income group
def createbins(x):
    if x<=70.00:
        return 'Low'
    elif x<=80.00:
        return 'Medium'
    else:
        return 'High'

credit_imp['INCOME_GROUP'] = credit_imp.NET_INCOME.apply(lambda x: createbins(x))
round(credit_imp.INCOME_GROUP.value_counts(normalize=True) * 100, 2)
```

```python
# Bar Plot of INCOME_GROUP to view binning

plt.figure(figsize = (15,7))

plt.subplot(121, title='NET_INCOME Histogram Distribution')
plt.xlabel('NET_INCOME')
plt.ylabel('FREQUENCY')
credit_imp.NET_INCOME.hist(ec='blue', bins=15)

plt.subplot(122, title='INCOME_GROUP Bar Plot')
plt.xlabel('INCOME_GROUP')
plt.ylabel('PERCENTAGE')
(credit_imp.INCOME_GROUP.value_counts(normalize=True)*100).plot.bar(rot=0)

plt.show()
```

```python
# PLotting heat maps to analyze Categorical to Categorical 

m_cols = ['CODE_GENDER','AGE_GROUP','INCOME_GROUP',
                    'NAME_CLIENT_TYPE', 'NAME_SELLER_INDUSTRY', 'CHANNEL_TYPE' ]

for i in range(len(m_cols)):
    col1 = m_cols[i]
    res = pd.pivot_table(data=merge_data, index=col1, columns= 'NAME_CONTRACT_STATUS', values='TARGET')
    plt.figure(figsize = (15,7))
    sns.heatmap(res, annot=True, cmap='YlGnBu')
    plt.subplots_adjust(top=0.8)
    plt.show()
```

## Clients who are more likely to have payment difficulties:
- Male clients having previous Contract status is refused
- Clients who are young have previous contract status is refused.
- Clients with low income group previous contract status is refused.
- Clients whose Client type is 'New ' and previous contract status as cancelled.
- Clients with'Seller Industry' as Jewelry and previous contract status as cancelled.
- Clients with channel type as 'AP + (Cash loans)' is refused.

## Clients who are less likely to have payment difficulties:

- Clients who are old have previous contract status as unused offer.
- Clients with client type as 'Refreshed' has previous contract status as approved.
- Clients with previous contract status as cancelled and 'Seller Industry' as Auto technology or Clothing or Construction indicated by zero
- Clients with previous contract status as unused offer and 'Seller Industry' as Auto technology or MLM Partners indicated by zero

```python
# Categorical - Numerical

c_cols = ['OCCUPATION_TYPE', 'TARGET']
n_cols = ['AMT_INCOME_TOTAL', 'DAYS_EMPLOYED']

for i in range(len(c_cols)):
    c_col = c_cols[i]
    for j in range(len(n_cols)):
        n_col = n_cols[j]
        sns.catplot(x= c_col, y = n_col, hue = 'NAME_CONTRACT_STATUS', kind = 'bar', data = merge_data, aspect=4)
        plt.xticks(rotation=90, fontweight = 'bold', fontsize = 25)
        plt.yscale('log')
        plt.title(c_col+ ' VS ' + n_col, fontweight = 'bold', fontsize= 20)
        plt.show()
```
